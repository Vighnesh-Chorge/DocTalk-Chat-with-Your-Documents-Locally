```python
# ─────────────────────────────────────────────
#  CELL 1 — Install
# ─────────────────────────────────────────────


# ─────────────────────────────────────────────
#  CELL 2 — Imports
# ─────────────────────────────────────────────
import fitz                   # PyMuPDF — reads PDF
import ollama                 # talks to Llama
import ipywidgets as widgets
from IPython.display import display, Markdown, clear_output


# ─────────────────────────────────────────────
#  CELL 3 — Extract text from PDF
# ─────────────────────────────────────────────
def extract_text(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text.strip()

pdf_path = input("Enter path to your PDF notes file: ").strip()
notes_text = extract_text(pdf_path)

print(f"✅ Extracted {len(notes_text)} characters from your notes.")
print("Preview:", notes_text[:300], "...")


# ─────────────────────────────────────────────
#  CELL 4 — Ask a question (with button UI)
# ─────────────────────────────────────────────
MODEL = "llama3.2"

SYSTEM_PROMPT = f"""You are a study assistant.
Answer ONLY from the student's notes below.
Write a clear exam-style answer in 7-10 lines.

--- NOTES ---
{notes_text[:12000]}
--- END ---
"""

while True:
    question = input("\nAsk a question (or type 'exit' to stop): ").strip()
    if question.lower() == "exit":
        break
    if not question:
        continue

    print("\nThinking...\n")
    response = ollama.chat(
        model=MODEL,
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user",   "content": question}
        ]
    )
    print("Answer:\n")
    print(response["message"]["content"])
    print("\n" + "="*60)
```
